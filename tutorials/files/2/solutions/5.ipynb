```python
imgPath="solutions/BaslerGrabbedImage-1551437682.png"
import cv2
import matplotlib.pyplot as plt
image = cv2.imread(imgPath)
angle90 = 90
(h, w) = image.shape[:2
                    ] 
# calculate the center of the image
center = (w / 2, h / 2)

cv2.cvtColor(image, cv2.COLOR_BGR2RGB, image)

M = cv2.getRotationMatrix2D((1920/2, 1080/2), angle90, 1) 
rotated90 = cv2.warpAffine(image, M, (1080, 1920)) 

plt.imshow(rotated90, interpolation='bilinear', origin='lower')
plt.show()
```
