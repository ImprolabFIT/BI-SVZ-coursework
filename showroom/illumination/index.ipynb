# Spektrum světla

Viditelné spektrum označuje rozsah elektromagnetického záření, které je viditelné lidským okem. Toto spektrum se dělí na různé vlnové délky, které vnímáme díky čípkům v lidském oku viz přednáška [Strojové vidění a fyzikální podstata](https://courses.fit.cvut.cz/BI-SVZ/lectures/files/bi-svz-01-strojove-videni-a-fyzikalni-podstata.pdf). Zdroje světla, jako je slunce a umělá světla, vyzařují různé vlnové délky v závislosti na jejich složení a teplotě.

Sluneční světlo, které je hlavním zdrojem světla pro většinu živých organismů na Zemi, vyzařuje spojité spektrum světla, které obsahuje všechny barvy duhy. Naproti tomu umělá světla, jako jsou LED diody, vyzařují díky svému specifickému složení a charakteristikám různá spektra světla.

Studená bílá LED světla, která se běžně používají v každodenním osvětlení, vyzařují světelné spektrum, které obsahuje všechny barvy duhy, ale je vychýleno směrem k modré části spektra. Je to proto, že bílá světla LED jsou obvykle vyrobena kombinací modré diody LED s fosforeskujícím materiálem, který vyzařuje žlutou barvu, což vede k vytvoření zdánlivě bílého světla.

Červená, zelená a modrá světla LED naopak vyzařují světlo, které je vychýlené směrem k příslušným barvám. Červená LED světla vyzařují světelné spektrum, které je převážně v červeném vlnovém rozsahu, zatímco zelená a modrá LED světla vyzařují světlo, které je převážně v zeleném a modrém vlnovém rozsahu.

# Monochromatické vs barevné kamery
### Barevná kamera je nezbytná při práci s aplikacemi, ve kterých je důležitá informace o barvě
Pokud aplikace potřebuje barevný obraz, je jedinou možností využití barevné kamery (například aplikace pro kontrolu a řízení kvality).

### Barevný senzor je ze své podstaty pomalejší
Zatímco u monochromatické kamery má každý pixel obvykle 8 bitů, u barevné kamery má každý pixel 8 bitů pro červený kanál, 8 bitů pro zelený kanál a 8 bitů pro modrý kanál, takže je třeba zpracovat třikrát více dat, což má za následek delší dobu zpracování a v důsledku toho i nižší snímkovací frekvenci. V mnoha případech machine vision aplikací je klíčová rychlost zpracování, proto je monochromatická kamera vhodnější.

### Monochromatická kamera dosahuje vyššího rozlišení
U barevné kamery může přítomnost **Bayerova filtru** přispět ke snížení dosažitelného rozlišení pořízených snímků. Pokud tedy není barevná informace nezbytná, měla by být upřednostněna monochromatická kamera. Rozdíly v rozlišení  jsou ukázány na obrázcích v přednášce [Strojové vidění a fyzikální podstata](https://courses.fit.cvut.cz/BI-SVZ/lectures/files/bi-svz-01-strojove-videni-a-fyzikalni-podstata.pdf).


### Spektrální odezva monochromatických a barevných kamer

Na obrázcích níže si můžete prohlédnout barevnou kameru **Basler ace AC 2500-60uc** a monochromatickou kameru **Basler ace AC 2500-60um**.

<img src="images/cameras.png" alt="Kamery" style="width:250px;"/>

Z obrázku, jenž ukazuje jejich spektrální odezvu je vidět, že monochromatická kamera má větší rozsah spektra, které je schopna zachytit. Barevná kamera má větší citlivost v oblasti viditelného spektra, ale v oblasti infračerveného spektra je citlivost nižší. Nižší citlivost v oblasti infračerveného spektra u barevných kamer je důsledek použití **IR (Infra Red) cut-off filtru** (více [informací](https://en.wikipedia.org/wiki/Infrared_cut-off_filter)), který je upevněn uvnitř patice objektivu. Použití filtru je možno vidět i na samotných kamerách viz. obrázek kamer výše. Tento fakt činí monochromatické kamery vhodné pro aplikace, kde potřebujeme pracovat s **NIR (Near Infra Red)** spektrem.

## Iluminační techniky

Podrobnější informace jsou k dispozici v přednášce [Kamerový systém a zpracování obrazu](https://courses.fit    .cvut.cz/BI-SVZ/lectures/files/bi-svz-03-kamerovy-system-a-zpracovani-obrazu.pdf).

### Osvětlení ve světlém poli (Bright field)

<img src="images/brightfield_pict.png"  alt="Osvětlení v temném poli"/>

Nejběžnější technika osvětlení nereflexních objektů.

### Osvětlení v tmavém poli (Dark field)

<img src="images/darkfield_pict.png"  alt="Osvětlení v temném poli"/>

Používá se ke zvýraznění reliéfu, či ke zvýraznění povrchových vad (škrábance atd.).

### Difuzní osvětlení 

<img src="images/diffuse_pict.png"  alt="Difuzní osvětlení"/>

Nejčastěji se používá u lesklých objektů nebo objektů se smíšenou reflexivitou, kde je zapotřebí rovnoměrné nasvícení objektu. Je k dispozici několik různých druhů světel, jenž difuzní osvětlení zajišťují. Příkladem jsou kupolovitá světla, či koaxiální světla.

### Zadní podsvícení (Back lighting)

<img src="images/back_pict.png"  alt="Zadní podsvícení"/>

Vytváří tmavé siluety objektů na světlém pozadí. Nejčastěji se používá ke zjišťování přítomnosti/nepřítomnosti otvorů/mezer, ke zjištění orientace objektů, či k jejich měření.
